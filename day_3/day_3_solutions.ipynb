```python
import jupyter_black

jupyter_black.load()

import numpy as np
from numpy.typing import NDArray
```

```python
with open("input_puzzle.txt") as f:
    input_digits = np.array([[j for j in s.strip()] for s in f.readlines()]).astype(int)
```

```python
%%time

result = (
    (
        input_digits[:, np.triu_indices(input_digits.shape[1], k=1)]
        * np.array([10, 1]).reshape(1, -1, 1)
    )
    .sum(axis=1)
    .max(axis=1)
    .sum()
)

print(f"Puzzle 1 Answer: {result}")
```

```python
def puzzle_2(input_digits: NDArray, k: int) -> NDArray:

    m, n = input_digits.shape

    base_indices = np.empty((m, k + 1), dtype=np.int64)
    base_indices[:, 0] = -1
    indices = base_indices[:, 1:]

    for r in range(m):
        for c in range(k):
            indices[r, c] = (
                input_digits[r, base_indices[r, c] + 1 : n - k + 1 + c].argmax()
                + base_indices[r, c]
                + 1
            )

    result = (
        input_digits[np.tile(np.arange(m).reshape(-1, 1), (1, k)), indices]
        * (10 ** np.arange(k)[::-1])
    ).sum()

    return result
```

```python
%%time

print(f"Puzzle 1 Answer: {puzzle_2(input_digits, 2)}")
```

```python
%%time

print(f"Puzzle 2 Answer: {puzzle_2(input_digits, 12)}")
```
