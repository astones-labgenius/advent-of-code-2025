```python
import numpy as np
from scipy.signal import convolve2d

import jupyter_black

jupyter_black.load()
```

```python
input_array = np.array(
    [
        [True if c == "@" else False for c in line]
        for line in np.loadtxt("puzzle_input.txt", dtype=str)
    ]
)
```

```python
%time

mask = np.ones((3, 3), dtype=int)
mask[1, 1] = 0

available_rolls = ((convolve2d(input_array, mask, mode="same") < 4) & input_array).sum()

print(f"Puzzle 1 Answer: {available_rolls}")
```

```python
%time

current_array = input_array.copy()
result_array = np.zeros_like(input_array).astype(bool)

remaining = True
while remaining:
    available_array = (convolve2d(current_array, mask, mode="same") < 4) & current_array
    result_array |= available_array
    current_array &= ~available_array
    remaining = available_array.sum() > 0

available_rolls = result_array.sum()
print(f"Puzzle 2 Answer: {available_rolls}")
```
