```python
import numpy as np
import re

from itertools import batched, chain
import jupyter_black

jupyter_black.load()
```

```python
with open("puzzle_input.txt") as f:
    range_dicts = [
        re.match(r"(?P<start>\d+)-(?P<end>\d+)", r).groupdict()
        for r in f.read().split(",")
    ]
```

```python
%%time


def test(i):
    i_str = str(i)
    len_i_str = len(i_str)
    return len_i_str % 2 == 0 and i_str[: len_i_str // 2] == i_str[len_i_str // 2 :]


running_sum = sum(
    chain.from_iterable(
        filter(test, range(int(d["start"]), int(d["end"]) + 1)) for d in range_dicts
    )
)

print(f"Puzzle 1 Answer: {running_sum}")
```

```python
%%time

result = sum(
    chain.from_iterable(
        filter(
            lambda i: any(
                len(set(batched(str(i), j + 1))) == 1
                for j in range(0, len(str(i)) // 2)
            ),
            range(int(d["start"]), int(d["end"]) + 1),
        )
        for d in range_dicts
    )
)

print(f"Puzzle 2 Answer: {result}")
```
