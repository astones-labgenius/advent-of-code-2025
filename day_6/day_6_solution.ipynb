```python
import jupyter_black
from itertools import pairwise
from collections import deque
from functools import reduce
import numpy as np

jupyter_black.load()
```

```python
with open("puzzle_input.txt") as f:
    lines = [s.strip() for s in f.readlines()]

column_operations = {i: c for i, c in enumerate(lines[-1]) if c.strip()}
```

```python
%%time


def parse_line(line: str, column_operations: list[str]) -> list[str, str]:
    parsed_line = []
    for i, j in pairwise(column_operations):
        parsed_line.append(line[i:j])
    parsed_line.append(line[j:])
    return parsed_line


parsed_numbers = np.array(
    [parse_line(l, column_operations) for l in lines[:-1]]
).astype(int)


result = sum(
    i.prod() if j == "*" else i.sum()
    for i, j in zip(parsed_numbers.T, column_operations.values())
)

print(f"Puzzle 1 Answer: {result}")
```

```python
with open("puzzle_input.txt") as f:
    lines = [s for s in f.readlines()]
```

```python
%%time

column_strings = np.array([([c for c in l]) for l in lines[:-1]])[:, :-1].T
joined_column_strings = ["".join(s).strip() for s in column_strings]

operations = deque(column_operations.values())
numbers = deque(joined_column_strings + [""])
c = 0
result = 0
while c < len(column_operations):
    terms = []
    while True:
        term = numbers.popleft()
        if not term:
            break
        terms.append(int(term))
    operation = operations.popleft()
    if operation == "*":

        result += reduce(lambda x, y: x * y, terms)
    else:
        result += sum(terms)
    c += 1

print(f"Puzzle 2 Answer: {result}")
```
