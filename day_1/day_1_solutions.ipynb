```python
from collections import Counter
import numpy as np

import jupyter_black

jupyter_black.load()
```

```python
with open("puzzle1_input.txt") as f:
    instructions = [f.strip() for f in f.readlines()]

instructions_integers = [(1 if i[0] == "R" else -1) * int(i[1:]) for i in instructions]
```

```python
state = np.empty(len(instructions_integers) + 1, dtype=int)
state[0] = 50
state[1:] = instructions_integers
state = state.cumsum()
```

```python
print(f"Puzzle 1 Answer: {Counter(state % 100)[0]}")
```

```python
expanded_state = np.empty(np.abs(instructions_integers).sum() + 1, dtype=int)
expanded_state[0] = 50
c = 0

for i in instructions_integers:
    if i >= 1:
        steps = np.arange(1, (i + 1))
    elif i <= 1:
        steps = np.arange(-1, i - 1, -1)
    else:
        raise ValueError
    end = c + abs(i)
    expanded_state[c + 1 : end + 1] = expanded_state[c] + steps
    c = end
```

```python
print(f"Puzzle 2 Answer: {Counter(expanded_state % 100)[0]}")
```
